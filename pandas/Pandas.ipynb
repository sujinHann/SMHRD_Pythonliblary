# DataFrame 병합
- concat
- merge

```python
# 데이터 생성
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# concat 1 axis=0
pd.concat([df1, df2, df3], axis = 0)
```

```python
# concat 2     axis=1
pd.concat([df1, df2, df3], axis = 1)
```

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'], 
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])
```

```python
df4
```

```python
# concat 4     df1,df4           axis=0
pd.concat([df1,df4], axis = 0)
```

```python
# concat 5     df1,df4           axis=1         인덱스 기준으로 병합
pd.concat([df1,df4], axis = 1)
```

```python
# join 속성 사용
# outer : 합집합. 기본값
pd.concat([df1,df4], axis = 1, join='outer')
```

```python
# inner : 교집합, 합치는 데이터들이 동일하게 가지고 있는 기준만 출력
pd.concat([df1,df4], axis = 1, join='inner')
```

```python
#merge 함수 
```

```python
#데이터 생성
df5 = pd.DataFrame({'key' : ['K0','K2','K3','K4'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})
df6 = pd.DataFrame({'key' : ['K0','K1','K2','K3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']})
```

```python
df6
```

```python
df6
```

```python
# 공통된 컬럼('key') 기준으로 merge
pd.merge(df5,df6,on='key')
```

```python
# how
# 병합의 방식을 지정
# concdt의 join과 동일한 기능
pd.merge(df5,df6,on='key', how='inner')
# inner : 공통된 데이터의 값만 출력(기본값)
# outer : 전체값 출력 
# left : 먼저 작은 데이터를 기준으로 값 출력
# right: 나중에 적은 데이터를 기준으로 값 출력 
```

```python
pd.merge(df5,df6,on='key', how='outer')
```

```python
pd.merge(df5,df6,on='key', how='left')
```

```python
df_merge = pd.merge(df5,df6,on='key', how='right')
```

```python
df_merge
```

```python
# 결측치 채우기
# fillna(채울값)
df_merge.fillna(0)
```

# 카테고리 생성하기

```python
# 데이터를 구간별로 구분하기
ages = [0, 2, 10, 21, 23, 37, 61, 20, 41,
       32, 100, 82, 58, 39, 77, 94, 16]
# 초과~이하
# 1~15(미성년자), 16~25(청년), 26~35(중년), 36~60(장년), 61~99(노년)
bins = [0, 15, 25, 35, 60, 99]
# label은 구간에 대한 이름
labels = ['미성년자', '청년', '중년', '장년', '노년']
```

```python
# cut() - 카테고리화 함수 
cats = pd.cut(ages, bins, labels = labels)
```

```python
cats
```

```python
# 각 항목별 갯수 확인 
cats.value_counts()
```

```python
age = pd.DataFrame(ages, columns=['나이'])
age
```
