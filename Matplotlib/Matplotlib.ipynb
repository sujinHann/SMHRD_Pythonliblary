```python
import numpy as np
```

```python
x=np.arange(7)
y=[1,4,5,8,9,5,3]
plt.plot(x,y,ls='--',linewidth=4 ,marker='o', ms=15, mfc='red', mec='green',mew='3')
plt.show()
```

```python
# 그림의 영역 지정
# xlim : x 범위 지정
# ylim : y 범위 지정
# xlim(최소 최대)
plt.plot(x,y,ls='--',linewidth=4 ,marker='o', ms=15, mfc='red', mec='green',mew='3')
plt.xlim(-2,8)
plt.ylim(-4,12)
plt.show()
```

```python
# grid 
# 격자, 모눈종이 모양

plt.plot(x,y,ls='--',linewidth=4 ,marker='o', ms=15, mfc='red', mec='green',mew='3')
plt.grid()
plt.show()
```

```python
# tick 설정
# xticks: : x축 틱 설정
# yticks : y축 틱 설정
plt.plot(x,y,ls='--',linewidth=4 ,marker='o', ms=15, mfc='red', mec='green',mew='3')
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.grid()
plt.show()
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x, color = 'r')
plt.plot(y, color = 'k')
plt.plot(z, color = 'b')
plt.show()
```

```python
plt.plot(x, color = 'r')
plt.show()
plt.plot(y, color = 'k')
plt.show()
plt.plot(z, color = 'b')
plt.show()
```

```python
# 범례 표시
plt.plot(x, label = 'x')
plt.plot(y, label = 'y')
plt.plot(z, label = 'z')
plt.legend(loc = 'right') #범례 띄우기
plt.show()

# best
#    upper right
#    upper left
#    lower left
#    lower right
#    right
#    center left
#    center right
#    lower center
#    upper center
#    center
```

```python
# 축에 타이틀 달기
# x축 이름 : xlabel
# y축 이름 : ylabel

plt.plot(x, label = '엑스')
plt.plot(y, label = '와이')
plt.plot(z, label = '제트')
plt.xlabel('x축')
plt.ylabel('y축',rotation = 0)
plt.title('x,y,z 그래프')
plt.legend(loc = 'right') #범례 띄우기
plt.show()
```

## 요일별 사망 교통사고 시각화

```python
data['요일']
```

```python
# 중복 없는 데이터 출력
data['요일'].unique()
```

```python
# 요일별 사고 발생 건수 확인하기
temp = data['요일'].value_counts()
```

```python
# 인덱싱
temp['월']
```

```python
# 월, 화, 수 인덱싱
temp['화']
```

```python
#월. 화, 수 가져오기
temp[['월','화','수']]
```

```python
# 월 ~ 일 가져오기
y = temp[['월','화','수','목','금','토','일']]
y
```

```python
x = y.index
x
```

```python
# barchart 그리기
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y)

plt.xlabel('요일')
plt.ylabel('사건건수',rotation=0)
plt.title('요일별 사망교통사고 건수')

plt.show()
```
